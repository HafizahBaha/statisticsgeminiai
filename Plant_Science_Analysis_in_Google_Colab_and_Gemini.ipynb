#STATISTICAL ANALYSIS

##ANALYSIS 1: INDEPENDENT SAMPLES t-TEST


RESEARCH QUESTION:
Is plant height different between the Control group and the High Fertilizer group?

 WHEN TO USE:
✅ Dependent variable is NUMERICAL (plant height)
✅ Comparing EXACTLY 2 INDEPENDENT groups




##ANALYSIS 2: ONE-WAY ANOVA


RESEARCH QUESTION:
Does fertilizer level (Control / Low / High) affect plant height?

WHEN TO USE:
✅ Dependent variable is NUMERICAL (plant height)
✅ Comparing 3 OR MORE INDEPENDENT groups


##ANALYSIS 3: PEARSON CORRELATION


RESEARCH QUESTION:
* Is there a relationship between plant height and fresh weight?
* Do taller plants also tend to weigh more?
WHEN TO USE:
* ✅ BOTH variables are NUMERICAL
* ✅ You want to know if they go up/down together
* ✅ NOT comparing groups — looking for a pattern


## ANALYSIS 4: CHI-SQUARE TEST

RESEARCH QUESTION:
* Is germination outcome (Germinated/Not Germinated) associated with treatment group (Control/Low/High Fertilizer)?

WHEN TO USE:
* ✅ BOTH variables are CATEGORICAL (labels/groups)
* ✅ You want to know if they are RELATED
